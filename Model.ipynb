```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('laptop_price.csv',encoding='latin-1')
```

```python
data.isnull().sum()
```

```python
data['Ram'] = data['Ram'].str.replace('GB','').astype('int64')
```

```python
data['Weight'] = data['Weight'].str.replace('kg','').astype('float64')
```

```python
data.head(2)
```

```python
data.info()
```

```python
len(data['Company'].value_counts())
```

```python
data['Company'].value_counts()
```

```python
def add_company(company_name):
    if (data['Company'] == company_name).sum()<10:
        return 'Other'
    else:
        return company_name

data['Company'] = data['Company'].apply(add_company)
```

```python
data['Company'].value_counts()
```

```python
data['Product'].value_counts()
```

```python
data['TypeName'].value_counts()
```

```python
data['ScreenResolution'].value_counts()
```

```python
data['TouchScreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['IPS'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data.sample(5)
```

```python
data['Cpu'].value_counts()
```

```python
data['Cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data['Cpu_name'].value_counts()
```

```python
def set_cpu(cpu):
    if cpu == 'Intel Core i7' or cpu == 'Intel Core i5' or cpu == 'Intel Core i3':
        return cpu
    else:
        if ('AMD' in cpu):
            return 'AMD'
        else:
            return 'Other'

data['Cpu_name'] = data['Cpu_name'].apply(set_cpu)
```

```python
data['Cpu_name'].value_counts()
```

```python
data.head()
```

```python
data['Gpu'].value_counts()
```

```python
def set_gpu(gpu):
    if 'Intel' in gpu:
        return 'Intel'
    elif 'Nvidia' in gpu:
        return 'Nvidia'
    elif 'AMD' in gpu:
        return 'AMD'
    else:
        return 'Other'

data['Gpu_name'] = data['Gpu'].apply(set_gpu)
```

```python
data['Gpu_name'].value_counts()
```

```python
data
```

```python
data = data[data['Gpu_name'] != 'Other']
```

```python
data.shape
```

```python
data['Gpu_name'].value_counts()
```

```python
data["OpSys"].value_counts()
```

```python
def set_os(os):
    if os in ['Windows 10','Windows 7','Windows 10 S']:
        return 'Windows'
    elif os=='Linux':
        return 'Linux'
    elif os=='Mac OS X' or os=='macOS':
        return 'Mac'
    else:
        return 'Other'

data.loc[:,'OpSys'] = data['OpSys'].apply(set_os)
```

```python
data['OpSys'].value_counts()
```

```python
data.head()
```

```python
data = data.drop(columns={'laptop_ID','Inches','Product','ScreenResolution','Cpu','Gpu'})
```

```python
data
```

```python
data.shape
```

```python
data = pd.get_dummies(data,dtype=int)
```

```python
data.head()
```

```python
data.shape
```

```python
x = data.drop('Price_euros',axis=1)
x.head()
```

```python
y = data['Price_euros']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
def get_model_accuracy(model):
    model.fit(x_train,y_train)
    acc = model.score(x_test,y_test)
    print(str(model) + ' : ' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
get_model_accuracy(lr)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
get_model_accuracy(lasso)
```

```python
from sklearn.tree import DecisionTreeRegressor
decision_tree = DecisionTreeRegressor()
get_model_accuracy(decision_tree)
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor()
get_model_accuracy(random_forest)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[25,50,100],
             'criterion':['squared_error', 'absolute_error', 'friedman_mse','poisson']}

grid_obj = GridSearchCV(estimator = random_forest,param_grid=parameters)

grid_fit = grid_obj.fit(x_train,y_train)

best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(x_test,y_test)
```

```python
import pickle
with open('lap_price_prediction.pickle','wb') as file:
    pickle.dump(best_model,file)
```
